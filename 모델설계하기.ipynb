```python
# 모두의 딥러닝 34쪽 참고


import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf

## 데이터 전처리 ## 
np.random.seed(3)
tf.random.set_seed(3)

df = pd.read_csv('ThoraricSurgery.csv')
# df.head()
df_1 = np.loadtxt('ThoraricSurgery.csv', delimiter=',')
train, test = train_test_split(df, test_size=0.3) ## 469*.7 >> 328.299999(학습데이터 70프로)

train_arr = train.to_numpy()
test_arr = test.to_numpy()
X_train = train_arr[:, 0:17]
Y_train = train_arr[:, 17]

## 모델 설계 ##
model = Sequential()
model.add(Dense(20, input_dim=17, activation='relu')) ## activation='softmax' 써보기 (더좋음)
model.add(Dense(20, activation='softmax')) ## 층(레이어)을 추가하기
model.add(Dense(1, activation='sigmoid')) ## Dense = 은닉층,  node 추가

## 모델 학습 ##
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, Y_train, epochs = 100, batch_size=10)



# print(type(train_arr))
# print(X_train)
# print(train.shape)

# X_train
```

```python
X_test = test_arr[:,0:17]
Y_test = test_arr[:,17]
## 모델 평가 ##
score = model.evaluate(X_test, Y_test, verbose = 0)
print(score)
```
